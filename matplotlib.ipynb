```python
import matplotlib.pyplot as plt
import numpy as np

X = range(100)
Y = [value**2 for value in X]
plt.plot(X,Y)
plt.show()
```

```python
X_1 = range(100)
Y_1 = [np.cos(value) for value in X]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X]

plt.plot(X_1,Y_1)
plt.plot(X_2,Y_2)
plt.show()
```

```python
fig = plt.figure() # figure 반환
fig.set_size_inches(10,5) # 크기지정
ax_1 = fig.add_subplot(1,2,1) # 두개의 plot 생성
ax_2 = fig.add_subplot(1,2,2) # 두 개의 plot 생성

ax_1.plot(X_1, Y_1, c='b') # 첫번째 plot
ax_2.plot(X_2, Y_2, c='g') # 두번째 plot
plt.show()
```

```python
X_1 = np.arange(100)
Y_1 = np.arange(100)

X_2 = np.arange(100)
Y_2 = np.arange(100) + 100

plt.plot(X_1, Y_1, color='#eeefff')
plt.plot(X_2, Y_2, color='r')
plt.show()
```

```python
plt.plot(X_1, Y_1, color='b', linestyle='dashed')
plt.plot(X_2, Y_2, color='r',ls='dotted')

plt.show()
```

```python
plt.plot(X_1, Y_1, color='b', linestyle='dashed')
plt.plot(X_2, Y_2, color='r',ls='dotted')

plt.title('Two lines')
plt.show()
```

```python
plt.plot(X_1, Y_1, color='b', linestyle='dashed',label='line_1')
plt.plot(X_2, Y_2, color='r',ls='dotted',label='line_2')

plt.title('y = ax + b')
plt.legend(shadow=True,fancybox=True,loc='lower right')
plt.show()
```

```python
plt.plot(X_1, Y_1, color='b', linestyle='dashed',label='line_1')
plt.plot(X_2, Y_2, color='r',ls='dotted',label='line_2')

plt.title('y = ax + b')
plt.grid(True,lw=0.4,ls='--',c='.90')
plt.legend(shadow=True,fancybox=True,loc='lower right')
plt.xlim(-100,200)
plt.ylim(-200,200)
plt.show()
```

```python
data = [[5., 25., 50., 20.],
        [4., 23., 51., 17],
        [6., 22., 52., 19]]

X = np.arange(0,8,2)
```

```python
plt.bar(X + 0.00, data[0], color = 'b', width = 0.50)
plt.bar(X + 0.50, data[1], color = 'g', width = 0.50)
plt.bar(X + 1.0, data[2], color = 'r', width = 0.50)
plt.xticks(X+0.50, ("A","B","C", "D"))
plt.show()
```

```python
color_list = ['b', 'g', 'r']
data_label = ["A","B","C"]
X = np.arange(np.array(data).shape[1])

for i in range(3):
    plt.bar(X, data[i], bottom = np.sum(data[:i], axis=0), 
            color = color_list[i], label=data_label[i])
plt.legend()
plt.show()
```

```python
import pandas as pd
```

```python
data = pd.read_csv('CMAPSS_sample.txt', sep=" ", header=None)
data.drop([26,27],axis =1, inplace = True)
```

```python
target_var = ['Target_Remaining_Useful_Life']
index_columns_names =  ["UnitNumber","Cycle"]
op_settings_columns = ["Op_Setting_"+str(i) for i in range(1,4)]
sensor_columns =["Sensor_"+str(i) for i in range(1,22)]

column_names = index_columns_names + op_settings_columns + sensor_columns
data.columns = column_names
```

```python
data.head()
```

```python
fig = plt.figure()
ax = []
for i in range(1,5):
    ax.append(fig.add_subplot(2,2,i))
ax[0].scatter(data['Sensor_15'],data['Cycle'])
ax[1].scatter(data['Sensor_16'],data['Cycle'])
ax[2].scatter(data['Sensor_17'],data['Cycle'])
ax[3].scatter(data['Sensor_18'],data['Cycle'])
plt.show()
```

```python
fig = plt.figure()
ax = []
for i in range(1,5):
    ax.append(fig.add_subplot(2,2,i))
ax[0].scatter(data['Sensor_15'],data['Cycle'],color='b',label = 'Sensor_15')
ax[1].scatter(data['Sensor_16'],data['Cycle'],color='g')
ax[2].scatter(data['Sensor_17'],data['Cycle'])
ax[3].scatter(data['Sensor_18'],data['Cycle'])

plt.subplots_adjust(wspace=0,hspace=0)
ax[0].legend()
ax[0].set_title('Sensor_15')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Exponential Distribution

```python
def exp_f(t, theta):
    return (1/theta) * np.exp(1) ** (-t/theta)
```

```python
t_list = np.linspace(0,5,100)
expf_list = [exp_f(x,0.5) for x in t_list]
```

```python
plt.plot(t_list,expf_list)
plt.show()
```

```python
expf_1 = [exp_f(x,0.5) for x in t_list]
expf_2 = [exp_f(x,1) for x in t_list]
expf_3 = [exp_f(x,2) for x in t_list]

plt.plot(t_list,expf_1,label='theta=0.5')
plt.plot(t_list,expf_2,label='theta=1')
plt.plot(t_list,expf_3,label='theta=2')
plt.grid(True,lw=0.4,ls='--',c='.90')
plt.legend()
plt.show()
```

```python
def exp_F(t, theta):
    return 1 - np.exp(1) ** (-t/theta)
```

```python
t = np.linspace(0,5,100)
expF = [exp_F(x,0.5) for x in t]
```

```python
plt.plot(t,expF)
plt.show()
```

```python
expF_1 = expF
expF_2 = [exp_F(x, 1) for x in t]
expF_3 = [exp_F(x, 2) for x in t]

plt.plot(t,expF_1,label='theta=0.5')
plt.plot(t,expF_2,label='theta=1')
plt.plot(t,expF_3,label='theta=2')
plt.grid(True,lw=0.4,ls='--',c='.90')
plt.legend()
plt.show()
```

# Weibull Distribution

```python
def Weib_F(t, beta, eta):
    return 1 - np.exp(-(t/eta)**beta)
```

```python
def Weib_f(t, beta, eta):
    return (beta/eta) * (t / eta) ** (beta-1) * np.exp(-(t/eta)**beta)
```

```python
t = np.linspace(0,5,100)
WeibF_1 = [Weib_F(x,0.8,1) for x in t]
WeibF_2 = [Weib_F(x, 1, 1) for x in t]
WeibF_3 = [Weib_F(x, 2, 1) for x in t]
WeibF_4 = [Weib_F(x, 2, 2) for x in t]

plt.plot(t, WeibF_1, label='beta=0.8 eta=1')
plt.plot(t, WeibF_2, label='beta=1 eta=1')
plt.plot(t, WeibF_3, label='beta=2 eta=1')
plt.plot(t, WeibF_4, label='beta=2 eta=2')
plt.grid(True, lw=0.4, ls='--', c='.90')
plt.title('Cumulative Weibull distribution plot')
plt.legend(shadow=True, fancybox=True,)
plt.show()
```

```python
t = np.linspace(0,5,100)
Weibf_1 = [Weib_f(x,0.8,1) for x in t]
Weibf_2 = [Weib_f(x, 1, 1) for x in t]
Weibf_3 = [Weib_f(x, 2, 1) for x in t]
Weibf_4 = [Weib_f(x, 2, 2) for x in t]

plt.plot(t, Weibf_1, label='beta=0.8 eta=1')
plt.plot(t, Weibf_2, label='beta=1 eta=1')
plt.plot(t, Weibf_3, label='beta=2 eta=1')
plt.plot(t, Weibf_4, label='beta=2 eta=2')
plt.grid(True, lw=0.4, ls='--', c='.90')
plt.title('Weibull distribution failure by parameter plot')
plt.legend(shadow=True, fancybox=True,)
plt.show()
```
